LendingClub is a US peer-to-peer lending company, headquartered in San Francisco, California.[3] It was the first peer-to-peer lender to register its offerings as securities with the Securities and Exchange Commission (SEC), and to offer loan trading on a secondary market. LendingClub is the world's largest peer-to-peer lending platform.

### Our Goal

Given historical data on loans given out with information on whether or not the borrower defaulted (charge-off), can we build a model that can predict wether or nor a borrower will pay back their loan? This way in the future when we get a new potential customer we can assess whether or not they are likely to pay back the loan. Keep in mind classification metrics when evaluating the performance of your model!

The "loan_status" column contains our label.

### Numerical features

```python
df[numerical_features].isnull().sum()
```

```python
df['revol_util'].value_counts().sort_index()
```

```python
df.groupby('loan_status')['revol_util'].mean()
```

```python
def impute_revo(cols):
    L = cols[0]
    r = cols[1]
    
    if pd.isnull(r) and L=='Charged Off':
        return 58
    elif pd.isnull(r) and L=="Fully Paid":
        return 52
    else:
        return r
```

```python
data = df.copy()
data['revol_util']=data[['loan_status','revol_util']].apply(impute_revo,axis=1)
```

```python
data['revol_util'].isnull().sum()
```

```python
df['revol_util'] = data['revol_util']
```

```python
df['pub_rec_bankruptcies'].value_counts()
```

```python
df.groupby('loan_status')['pub_rec_bankruptcies'].mean()
```

```python
df['pub_rec_bankruptcies'].fillna(0.0,inplace=True)
```

```python
df['mort_acc'].value_counts()
```

```python
df.groupby('loan_status')['mort_acc'].mean()
```

```python
import random
def impute_mort(acc):
    L=[0.0,1.0,2.0,3.0]
    if pd.isnull(acc):
        return random.choice(L)
    return acc
```

```python
df['mort_acc'] = df['mort_acc'].apply(impute_mort)
```

```python
nominal = ['term__ 60 months','home_ownership__OWN','home_ownership__RARE','home_ownership__RENT','initial_list_status__w']
```

```python
X= df[selected]
y= df['loan_status']
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Model Selection

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()

model.add(Dense(19,'sigmoid'))
model.add(Dense(19,'sigmoid'))
model.add(Dense(19,'sigmoid'))
model.add(Dense(1,'sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
Early_Stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x=X_train.values,y=y_train.values,validation_data=(X_test.values,y_test.values),
         batch_size=512,callbacks=Early_Stop,epochs=1200,verbose=1)
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
pred = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred))
```
